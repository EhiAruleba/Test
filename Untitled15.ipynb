### Preprocess data

```python
#scale data
data = data.map(lambda x, y: (x/255, y))
```

```python
scaled_iterator = data.as_numpy_iterator()
```

```python
batch = scaled_iterator.next()
```

```python
len(data)
```

```python
train_size = int(len(data)*.7)
vald_data_size = int(len(data)*.2)
test_data_size = int(len(data)*.1)
```

```python
train_size
```

```python
train = data.take(train_size)
vald = data.skip(train_size).take(vald_data_size)
test = data.skip(train_size+vald_data_size).take(test_data_size)
```

```python
len(test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model =Sequential()
model.add(Conv2D(16, (3,3), 1, activation ='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation = 'relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation = 'relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile('adam', loss =tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
#Train Model
logdir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs = 20, validation_data = vald, callbacks=[tensorboard_callback])
```

```python
#evaluate model
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
Pre =Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    y2 = model.predict(X)
    Pre.update_state(y, y2)
    re.update_state(y, y2)
    acc.update_state(y, y2)
```

```python
print(Pre.result(), re.result(), acc.result())
```
